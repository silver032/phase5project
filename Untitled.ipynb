```python
# Import necessary libraries
import pandas as pd

# Load the datasets
holidays_events = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\holidays_events.csv")
oil = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\oil.csv")
stores = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\stores.csv")
test = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\test.csv")
train = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\train.csv")
transactions = pd.read_csv(r"C:\Users\silve\Downloads\store-sales-time-series-forecasting\transactions.csv")

# Display the first few rows of each dataset to understand their structure
print("Holidays Events Data:")
print(holidays_events.head(), "\n")

print("Oil Data:")
print(oil.head(), "\n")

print("Stores Data:")
print(stores.head(), "\n")

print("Test Data:")
print(test.head(), "\n")

print("Train Data:")
print(train.head(), "\n")

print("Transactions Data:")
print(transactions.head(), "\n")
```

```python
# Summary statistics for numerical columns in each dataset
print("Summary Statistics for Holidays Events Data:")
print(holidays_events.describe(), "\n")

print("Summary Statistics for Oil Data:")
print(oil.describe(), "\n")

print("Summary Statistics for Stores Data:")
print(stores.describe(), "\n")

print("Summary Statistics for Test Data:")
print(test.describe(), "\n")

print("Summary Statistics for Train Data:")
print(train.describe(), "\n")

print("Summary Statistics for Transactions Data:")
print(transactions.describe(), "\n")
```

```python
# Check for missing values in each dataset and document the findings

# Holidays Events
print("Missing Values in Holidays Events Data:")
print(holidays_events.isnull().sum(), "\n")

# Oil Data
print("Missing Values in Oil Data:")
print(oil.isnull().sum(), "\n")

# Stores Data
print("Missing Values in Stores Data:")
print(stores.isnull().sum(), "\n")

# Test Data
print("Missing Values in Test Data:")
print(test.isnull().sum(), "\n")

# Train Data
print("Missing Values in Train Data:")
print(train.isnull().sum(), "\n")

# Transactions Data
print("Missing Values in Transactions Data:")
print(transactions.isnull().sum(), "\n")
```

```python
# Merge train data with store data
train_store_merged = pd.merge(train, stores, on='store_nbr', how='left')

# Display the first few rows to understand the relationship
print("Merged Train and Stores Data:")
print(train_store_merged.head(), "\n")
```

```python
# Merge train data with oil data
train_oil_merged = pd.merge(train, oil, on='date', how='left')

# Display the first few rows to understand the relationship
print("Merged Train and Oil Data:")
print(train_oil_merged.head(), "\n")
```

```python
# Merge train data with holidays_events data
train_holidays_merged = pd.merge(train, holidays_events, on='date', how='left')

# Display the first few rows to understand the relationship
print("Merged Train and Holidays Events Data:")
print(train_holidays_merged.head(), "\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to visualize sales by store type
plt.figure(figsize=(10, 6))
sns.boxplot(x='type', y='sales', data=train_store_merged)
plt.title("Sales by Store Type")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Scale sales and oil prices
train_oil_merged['sales_scaled'] = scaler.fit_transform(train_oil_merged[['sales']])
train_oil_merged['oil_scaled'] = scaler.fit_transform(train_oil_merged[['dcoilwtico']])
```

```python
# Plot scaled sales and oil prices over time
plt.figure(figsize=(14, 7))
plt.plot(train_oil_merged['date'], train_oil_merged['sales_scaled'], label='Sales (scaled)')
plt.plot(train_oil_merged['date'], train_oil_merged['oil_scaled'], label='Oil Price (scaled)', alpha=0.7)
plt.title("Sales vs Oil Prices Over Time (Scaled)")
plt.xlabel("Date")
plt.ylabel("Scaled Value")
plt.legend()
plt.show()
```

```python
# Calculate average sales before, during, and after holidays
average_sales_holidays = train_holidays_merged.groupby(['type'])['sales'].mean().reset_index()

# Bar plot to visualize average sales during different holiday types
plt.figure(figsize=(12, 6))
sns.barplot(x='type', y='sales', data=average_sales_holidays)
plt.title("Average Sales by Holiday Type")
plt.show()
```

```python
# Calculate correlations in the merged dataset (e.g., train and oil)
correlation_matrix = train_oil_merged[['sales', 'dcoilwtico']].corr()

# Display the correlation matrix
print("Correlation Matrix between Sales and Oil Prices:")
print(correlation_matrix, "\n")
```

```python
# Merge train data with store data, oil data, transactions, and holidays events
train_combined = pd.merge(train, stores, on='store_nbr', how='left')
train_combined = pd.merge(train_combined, oil, on='date', how='left')
train_combined = pd.merge(train_combined, transactions, on=['store_nbr', 'date'], how='left')
train_combined = pd.merge(train_combined, holidays_events, on='date', how='left')

# Display the first few rows to confirm the merge
print(train_combined.head(), "\n")
```

```python
# Select only numerical columns for correlation analysis
numerical_columns = train_combined.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numerical columns
correlation_matrix = numerical_columns.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix, "\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Add title and labels
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
# Correlation of sales with other variables
sales_correlations = correlation_matrix['sales'].sort_values(ascending=False)
print("Correlation of Sales with Other Variables:")
print(sales_correlations, "\n")
```
